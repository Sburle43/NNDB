```python
import os
import csv
import pandas as pd
import html5lib
# from schema import SCHEMA
import logging
import logging.config
import selenium
from selenium import webdriver
import time
from PIL import Image
import io
import requests
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By

```

```python
# #Install driver


opts=webdriver.ChromeOptions()
opts.headless=True

driver = webdriver.Chrome(ChromeDriverManager().install() ,options=opts)
```

```python
#links for A, B, C,...Z


driver.get("https://www.nndb.com/lists/503/000063314/")
time.sleep(2)
```

```python
#calculate the rows 

rows = 1+len(driver.find_elements(By.XPATH,"/html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/p[6]/table/tbody/tr"))
print(rows)
# # Read and Convert Web table into data frame

# cols = len(driver.find_elements(By.XPATH,"/html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/p[6]/table/tbody/tr[1]/td"))
# print(cols)
```

```python
# import csv

#code to document the names, occupation and href links
f = open('K.csv', 'w')
writer = csv.writer(f)
header=['name','Occ', 'href']
writer.writerow(header)
f.close()

for r in range(1, rows):
    f = open('K.csv', 'a')
    writer = csv.writer(f)
    

    l1=[]
    name = driver.find_element(By.XPATH,
            "//html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/p[6]/table/tbody/tr["+str(r)+"]/td[1]").text

    occ = driver.find_element(By.XPATH,
            "//html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/p[6]/table/tbody/tr["+str(r)+"]/td[2]").text
    href = driver.find_element(By.XPATH,
            "//html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/p[6]/table/tbody/tr["+str(r)+"]/td[1]/a").get_attribute("href")
    
    l1.append(name)
    l1.append(occ)
    l1.append(href)
#     l1.append(gender_details)
    writer.writerow(l1)
    f.close()
    print(l1)
driver.close()
    
```

```python
def get_details(href):
    driver.get(href)
    
    try:
        name = driver.find_element(By.XPATH, "/html/body/center/table/tbody/tr/td[1]/table/tbody/tr[3]/td/table/tbody/tr/td/font").text
    
    except:
        name =None
    
    try:
        born =driver.find_element(By.XPATH,"//b[text()= 'Born:']/parent::p").text.splitlines()[0:2]
    except:
        born = None
    
    try:
        gender_details = driver.find_element(By.XPATH,"//b[text()= 'Gender:']/parent::p").text.splitlines()

    except:
        gender_details = None
        
    try:
        
        basic = driver.find_element(By.XPATH,"//b[text()= 'Brother:']/parent::p | //b[text() = 'Wife:']/parent::p |//b[text() = 'Father:']/parent::p|//b[text() = 'Mother:']/parent::p | //b[text() = 'Son:']/parent::p | //b[text() = 'Daughter:']/parent::p").text.splitlines()
    except:
        basic = None
    
    return [name,born,gender_details,basic]
```
